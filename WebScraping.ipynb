The purpose of this notebook is to work on scraping web sites.

```python
import urllib
import requests
from bs4 import BeautifulSoup
```

```python
oregon = 'http://www.oregonbids.com'
r = requests.get(oregon)
soup = BeautifulSoup(r.text, "lxml")
print(soup.prettify())
```

```python
#Get title of page
soup.title
```

```python
# Finds all links

all_links=soup.find_all("a")

for link in all_links:
    print(link.get("href"))
```

```python
# Find all tables
all_tables = soup.find('table')
```

Parse all table data and create a dataframe

```python
A = []
B = []
C = []
D = []

for row in all_tables.find_all('tr'): 
    cells = row.find_all('td')
    if len(row)==int(3):
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
    elif len(row)==int(1):
        S = BeautifulSoup(str(cells[0]), "lxml").text
        if len(S)>int(1):
            if S.find('Scope')!= -1:
                D.append(S)
   
```

```python
import pandas as pd
df=pd.DataFrame(A[1:], columns=['Due Date'])
df['Solicitation Title'] = B[1:]
df['Organization'] = C[1:]
df['Scope'] = D
```
